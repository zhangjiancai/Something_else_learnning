```python
from sklearn.datasets import make_moons
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data, target = make_moons()

df = pd.DataFrame(data)
df['target'] = target

df[0].plot(kind='kde',secondary_y=True,label='df[0]')   # 分布完全重合，所以只有一条线
df[1].plot(kind='kde',secondary_y=True,label='df[1]')
plt.legend()

df1 = df[df['target']==0]
df2 = df[df['target']==1]

#df1.index = range(len(df1))    # 调整行索引（养成良好的数据处理习惯）
#df2.index = range(len(df2))

plt.figure(figsize=(3,3))
plt.scatter(df1[0],df1[1],color='red')
plt.scatter(df2[0],df2[1],color='green')


```

```python
import random
#创建一个小时内温度随时间变化的曲线（以分钟计算，温度在15到20之间）
#1.准备数据
x = range(60)
y = [random.uniform(15,20)for i in x]

#2.创建画布
plt.figure( figsize=(12,8), dpi=80 )
plt.plot(x,y)

#3 设置刻度及步长
z = range(40)
x_label = ['11:{}'.format(i) for i in x]
plt.xticks( x[::5], x_label[::5])
plt.yticks(z[::5])  #5是步长

#4 添加网格信息
plt.grid(True, linestyle='--', alpha=0.5) #默认是True，风格设置为虚线，alpha为透明度

#5 添加标题（中文在plt中默认乱码，不乱码的方法在本文最后说明）
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Curve of Temperature Change with Time')

#6 保存图片，并展示
#plt.savefig(".\\plt_png\test1.2.png")
#plt.show()
```
