```python
import pandas as pd
from pandas import DataFrame
```

```python
from sklearn import datasets
```

自动生成数据

```python
centers = [[1,1],[-1,1],[0,0],[1,0]]
x,labels_truex = datasets.make_blobs(n_samples=750,n_features=4,centers=centers,cluster_std=[0.4,0.2,0.2,0.4],random_state=10) 
```

```python
y,labels_truey = datasets.make_circles(n_samples=1000,factor=0.2,noise=0.05)
z,labels_truez = datasets.make_moons(n_samples=1000,noise=0.05)
```

K-means

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
X = np.concatenate((x,y,z))
datavisable(X,"blue")
```

```python
def KMeans_data(data,n):

    k_means = KMeans(n_clusters=n, random_state=10)

    k_means.fit(data)

    labels_true_predict = k_means.predict(data)
    return labels_true_predict
```

```python
##print(KMeans_data(X,4))
datavisable(X,KMeans_data(X,4))
```

```python
datavisable(x,KMeans_data(x,4))
```
