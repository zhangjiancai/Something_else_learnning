```python
import pandas as pd
from pandas import DataFrame
```

```python
from sklearn import datasets
```

自动生成数据

```python
centers = [[1,1],[-1,1]]
x,labels_truex = datasets.make_blobs(n_samples=750,centers=centers,cluster_std=0.4,random_state=0)  # type: ignore
```

```python
y,labels_truey = datasets.make_circles(n_samples=1000,factor=0.2,noise=0.05)
z,labels_truez = datasets.make_moons(n_samples=1000,noise=0.05)
```

数据可视化

```python
import matplotlib.pyplot as plt
```

```python
def datavisable(data,target):
    df = pd.DataFrame(data)
    df['target'] = target
    df1 = df[df['target']==0]
    df2 = df[df['target']==1]
    df1.index = range(len(df1))    # type: ignore # 调整行索引（养成良好的数据处理习惯）
    df2.index = range(len(df2))  # type: ignore
    plt.figure(figsize=(3,3))
    plt.scatter(df1[0],df1[1],color='red')
    plt.scatter(df2[0],df2[1],color='green')
```

```python
datavisable(x,labels_truex)
```

```python
datavisable(y,labels_truey)
```

```python
datavisable(z,labels_truez)
```
